```
file_name_list = []
verification = []
rawtest_speaker_list = []
rawtest_speaker = []

len_total = []
len_agent = []
len_customer = []
f_name = []
call_duration = []

for files in os.listdir('/content/ICICI Call Summary/'):
  file_name = '/content/ICICI Call Summary/'+files
  f = open(file_name)

  data = json.load(f)
  count_total = 0
  count_agent = 0
  count_customer = 0
  for i in data['SNIPPETS']:
    count_total+=1
    # print('rawText',i['rawText'])
    # print('speaker',i['speaker'])
    file_name_list.append(files)
    verification.append(' ')
    file_name_list.append(files)
    verification.append(' ')
    rawtest_speaker_list.append('raw_test')
    rawtest_speaker.append(i['rawText'])
    rawtest_speaker_list.append('speaker')
    rawtest_speaker.append(i['speaker'])
    if(i['speaker'] == 'agent'):
      count_agent+=1
    else:
      count_customer+=1



  # print(files)
  f_name.append(files)
  # print(count)
  len_total.append(count_total)
  len_agent.append(count_agent)
  len_customer.append(count_customer)

  start_time_str = data['START_TIME'][:-1]
  end_time_str = data['END_TIME'][:-1]
  start_time = datetime.datetime.fromisoformat(start_time_str)
  end_time  = datetime.datetime.fromisoformat(end_time_str)
  duration_in_seconds = (end_time - start_time).total_seconds()
  call_duration.append(duration_in_seconds)

  f.close()
```

```
final_df = pd.DataFrame()

final_df['file_name'] = file_name_list
final_df['verification'] =verification
final_df['type'] = rawtest_speaker_list
final_df['value'] =rawtest_speaker
```

```
len_df = pd.DataFrame()

len_df['file_name'] = f_name
len_df['total_length'] = len_total
len_df['agent_length'] = len_agent
len_df['customer_length'] = len_customer

call_dur1 = []
for i in call_duration:
  call_dur1.append(i/60)

len_df['call_duration(in minutes)'] = call_dur1
```
